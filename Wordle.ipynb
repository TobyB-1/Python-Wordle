```python
guesses: int = 1 # this will track the guesses (important for the game loop)
running: bool = True # this will help with the game loop
guess_list: list = [] # track guessed words
row: list = [] # create a row for the squares
```

👆Here is the main variables (that dont need to be inside the loop)


*   The 'guess_list' variable will make sure if they play again they are able **to guess words they guessed in the last round**
*   The 'row' will **store the colours for the guess**





```python
def runLoop():
    choice = str(input("Play again? (Yes or No): "))
    # check the input is valid
    while choice not in ["Yes", "No"]:
       choice = str(input("Play again? (Yes or No): "))
    # if they say yes then keep the loop, if no then exit
    if choice == "Yes":
        global guesses
        guesses = 1
        global running
        running = True
    else:
        sys.exit()
```

👆Here i create a function to manage the game loop, so that if the user wishes to play again, **keep the loop going**
*   'sys.exit()' - this makes use of the import i made above (it shuts down the program)



```python
while running:
    if guesses < 7:
        if guesses == 1:
            global word
            word = random.choice(word_list) # generate a word
        print(f"Guess: {guesses}")
        guess = str(input("Type in a 5 letter word: "))
        # check the input is valid
        while len(guess) != 5 or guess not in word_list or guess in guess_list:
            if len(guess) != 5:
                print("Word is not 5 letters!")
                guess = str(input("Type in a 5 letter word: "))
            elif guess in guess_list:
                print("Word already guessed!")
                guess = str(input("Type in a 5 letter word: "))
            else:
                print("Word not found!")
                guess = str(input("Type in a 5 letter word: "))
        guess_list.append(guess) # add the guess to the list, so if they say the same word they are prompted again
        print(guess)
        # loop through each letter in the word
        for x in range(5):
            if guess[x] == word[x]:
                row.append("🟩") # add the square that is resulted to the list
            elif guess[x] in word:
                row.append("🟨")
            else:
                row.append("⬛")
        print(row)
        # if the guess the word then they win
        if guess == word:
            print("You win!")
            runLoop()
        guesses += 1
    # if they have ran out of guesses and they have not got the word then they lose
    if guesses == 7 and guess != word:
        print(f"You lost! Word was, {word}")
        guess_list.clear() # clear the list so that they can guess the same words next round
        runLoop()
```

👆Here is the main loop:

*   Line #1 - this while loop is **the main game loop**
*   Line #2 - this is the inner loop that makes sure the user **only has 6 attempts**
*   Lines #3-5 - at the start of the loop, this checks if the game has just started by **checking the guess amount**, i have done this so that a new **different** word can be generated at the start of every round
*   Lines #9-18 - this whole loop checks the input from 'guess' to make sure **its valid**, the conditions are: **word must be 5 letters**, **word must not have been guessed already** (here is the reason i created the 'guess_list') & the word **must be in the index** (word_list)
